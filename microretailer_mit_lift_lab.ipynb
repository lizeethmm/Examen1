```python
#Se eliminara los outliers a trav√©s del metodo de desviaci√≥n estandar. 
y=data4
Limite_superior=y.mean()+3*y.std()
Limite_inferior=y.mean()-3*y.std()
print('Limite superior permitido',Limite_superior)
print('Limite inferior permitido',Limite_inferior)
```

```python
#Encontramos outliers del dataframe
outliers= data4[(y>Limite_superior)|(y>Limite_inferior)]
outliers
```

```python
#Obtenemos datos sin outliers del dataframe. DATOS LIMPIOS
data5=data4[(y<=Limite_superior)&(y>=Limite_inferior)]
data5
```

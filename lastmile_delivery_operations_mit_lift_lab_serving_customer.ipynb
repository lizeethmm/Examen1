```python
#Separamos valores cuantitativos 
datos=data.iloc[ : , [4,9,11,12,14,15,16,19,20,24]]
datos
```

```python
#Corroboramos valores nulos
valores_nulos=datos.isnull().sum()
valores_nulos
#Se observa O datos nulos
```

```python
#Realizamos una copia del dataframe
data2=datos.copy()
```

```python
#Se eliminara los outliers a través del metodo de desviación estandar. 
y=data2
Limite_superior=y.mean()+3*y.std()
Limite_inferior=y.mean()-3*y.std()
print('Limite superior permitido',Limite_superior)
print('Limite inferior permitido',Limite_inferior)
```

```python
#Encontramos outliers del dataframe
outliers= data2[(y>Limite_superior)|(y>Limite_inferior)]
outliers
```

```python
#Obtenemos datos sin outliers del dataframe. DATOS LIMPIOS
data3=data2[(y<=Limite_superior)&(y>=Limite_inferior)]
data3
```

```python
#Corroboramos valores nulos
valores=data3.isnull().sum()
valores
```

```python
#Eliminimamoas valores nulos del dataframe con mean. Es decir a traves de la función promedio
data4=data3.fillna(round(data3.mean(),1))
data4

```

```python
#corroboramos valores nulos del dataframe 
data_clean=data4.isnull().sum()
data_clean
```

```python
#Aplicar Filtro por columnas para encontrar valores cualitativos
datos2=data.iloc[ : , [0,1,2,3,5,6,7,8,10,13,17,18,21,22,23,25,26]]
datos2.info()
```

```python
#Corroboramos valores nulos
data_clean1=datos2.isnull().sum()
data_clean1
```

```python
#Limpieza de columnas cualitativas con no informacion
data8= datos2.fillna('no informacion')
data8
```

```python
#Unión de columnas cualitativas y cuantitativas
Datos_limpios= pd.concat([datos2,data_clean], axis=1)
Datos_limpios
```

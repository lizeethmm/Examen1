```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(12, 6))
plt.boxplot(data["238_frequency_buying_store"]) #Dato númerico
plt.title("Outliers de Frecuencia de Venta en la Tienda")
plt.show() #Se dibuja el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75. Para tener mejor precisión y sutil. Ya que hay menor rango.
y=data1["238_frequency_buying_store"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos de "81_number_online_purchases_month"
y1=data1["81_number_online_purchases_month"]
Limite_Superior1= y1.mean() + 3*y1.std()
Limite_Inferior1= y1.mean() - 3*y1.std()
print("Limite superior permitido", Limite_Superior1)
print("Limite inferior permitido", Limite_Inferior1)
```
